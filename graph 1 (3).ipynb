# graphs of function

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
x=np.linspace(-10,10,15)
y=x**2
```

```python
x
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x_axis',fontsize=10)
plt.ylabel('y_axis',fontsize=10)
```

```python

x=np.linspace(-50,100,5)
y=x-np.exp(x)
```

```python
x
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x_axis',fontsize=5)
plt.ylabel('y_axis',fontsize=5)
```
